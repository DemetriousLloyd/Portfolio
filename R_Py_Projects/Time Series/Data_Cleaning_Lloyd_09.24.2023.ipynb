# Dataframe Creation and Inspection

## Change the file locations to fit your code

```python
globe = 'C:/Users/raven/Desktop/Temp_GitHub_Collab/archive/GlobalTemperatures.csv'
city = 'C:/Users/raven/Desktop/Temp_GitHub_Collab/archive/GlobalLandTemperaturesByCity.csv'
country = 'C:/Users/raven/Desktop/Temp_GitHub_Collab/archive/GlobalLandTemperaturesByCountry.csv'
major_city = 'C:/Users/raven/Desktop/Temp_GitHub_Collab/archive/GlobalLandTemperaturesByMajorCity.csv'
state = 'C:/Users/raven/Desktop/Temp_GitHub_Collab/archive/GlobalLandTemperaturesByState.csv'

temp_global = pd.read_csv(globe)
temp_city = pd.read_csv(city)
temp_country = pd.read_csv(country)
temp_major_city = pd.read_csv(major_city)
temp_state = pd.read_csv(state)

print(temp_global.info())
print(temp_city.info())
print(temp_country.info())
print(temp_major_city.info())
print(temp_state.info())
```

# 'dt' column name was annoying me. Renaming columns to 'date'

```python
temp_global = temp_global.rename(columns={'dt' : 'date'})
temp_city = temp_city.rename(columns={'dt' : 'date'})
temp_country = temp_country.rename(columns={'dt' : 'date'})
temp_major_city = temp_major_city.rename(columns={'dt' : 'date'})
temp_state = temp_state.rename(columns={'dt' : 'date'})
```

# Null Value Observations
## Null Values appear as missing (white or blank) in the shadow matrix

```python
msno.matrix(temp_global, labels = True)
msno.matrix(temp_city, labels = True)
msno.matrix(temp_country, labels = True)
msno.matrix(temp_major_city, labels = True)
msno.matrix(temp_state, labels = True)
```

# Drop Rows with Missing Values. 
## Use heatmaps to determine how missing values were correlated

```python
# temp_global
msno.heatmap(temp_global, cmap ='rainbow')
temp_global = temp_global.dropna()
```

```python
# temp_city

msno.heatmap(temp_city, cmap ='rainbow')
temp_city = temp_city.dropna()

```

```python
# temp_country

msno.heatmap(temp_country, cmap ='rainbow')
temp_country = temp_country.dropna()

```

```python
# temp_major_city

msno.heatmap(temp_major_city, cmap ='rainbow')
temp_major_city = temp_major_city.dropna()

```

```python
# temp_state

msno.heatmap(temp_state, cmap ='rainbow')
temp_state = temp_state.dropna()
```
